```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import string
import numpy as np
import pandas as pd
import jieba
```

```python
# Load the text data from the txt file
with open('death reasondo.txt', 'r') as f:
    text_data = f.readlines()
    text_string=''.join(text_data)

texts=text_string.replace('"',"").split(" ")

def preprocess(text):
    # remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    # tokenize the text using jieba
    tokens = jieba.cut(text)
    # remove stop words
    stop_words = ['的', '了', '是', '我', '你', '他']
    tokens = [token for token in tokens if token not in stop_words]
    # join the tokens back into a string
    text = ' '.join(tokens)
    return text
preprocessed_texts_new = [preprocess(text) for text in texts]
print(preprocessed_texts_new)
vectorizer = CountVectorizer()
X_new = vectorizer.fit_transform(preprocessed_texts_new)
print(X_new.shape)
print(X_new)
```
